```python

import matplotlib.pyplot as plt
import numpy as np


# prepare some coordinates
x, y, z = np.indices((8, 8, 8))

# draw cuboids in the top left and bottom right corners, and a link between
# them
cube1 = (x < 3) & (y < 3) & (z < 3)
cube2 = (x >= 5) & (y >= 5) & (z >= 5)
link = abs(x - y) + abs(y - z) + abs(z - x) <= 2

# combine the objects into a single boolean array
voxelarray = cube1 | cube2 | link

# set the colors of each object
colors = np.empty(voxelarray.shape, dtype=object)
colors[link] = 'red'
colors[cube1] = 'blue'
colors[cube2] = 'green'

# and plot everything
ax = plt.figure().add_subplot(projection='3d')
ax.voxels(voxelarray, facecolors=colors, edgecolor='k')

plt.show()
```

```python
from matplotlib import pyplot as plt

class LineBuilder:
    def __init__(self, line):
        self.line = line
        self.xs = list(line.get_xdata())
        self.ys = list(line.get_ydata())
        self.cid = line.figure.canvas.mpl_connect('button_press_event', self)

    def __call__(self, event):
        print('click', event)
        if event.inaxes!=self.line.axes: return
        self.xs.append(event.xdata)
        self.ys.append(event.ydata)
        self.line.set_data(self.xs, self.ys)
        self.line.figure.canvas.draw()

fig, ax = plt.subplots()
ax.set_title('click to build line segments')
line, = ax.plot([0], [0])  # empty line
linebuilder = LineBuilder(line)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math

def pert(a, b, c, *, size=1, lamb=4):
    r = c - a
    alpha = 1 + lamb * (b - a) / r
    beta = 1 + lamb * (c - b) / r
    return a + np.random.beta(alpha, beta, size=size) * r

arr = pert(48, 48, 71, size=10_000)
arr_rounded = list(map(math.floor,arr))
print(arr_rounded)
print(arr)
plt.hist(arr)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math

def even_distribution(min,max,n_items):
    n_bins = (max-min)+1
    items_per_bin = int(math.floor(n_items/n_bins))

    final_array = []
    for x in range(min,max+1):
        final_array+=[x]*items_per_bin
    
    return final_array

a = even_distribution(0,20,10000)
print(len(a))
plt.hist(a,bins=21)


```

```python
import matplotlib.pyplot as plt

def linear_etd_function(etd_difference):
    #y = mx+b
    #b always 0.5 (difference of zero = 0.5)
    # 0 at - (max synthetic etd)
    # 1 at max synthetic_etd
    # slope is therefore 0.5/max_synthetic etd
    max_synthetic_etd = 200

    return ((0.5/max_synthetic_etd)*etd_difference)+0.5

etds = [-200, 0, 200]

etds_linearized = [linear_etd_function(etd) for etd in etds]

fig,ax = plt.subplots()

ax.plot(etds,etds_linearized,color = 'darkorchid',linewidth = 3)
ax.set_xlabel("ETD difference (steps)")
ax.set_ylabel("ETD value")
ax.set_title("ETD value as a function of ETD difference")
ax.set_yticks([0.5], minor=True)
ax.set_xticks([0], minor=True)
ax.yaxis.grid(True, linewidth = 1, color = 'black' ,which='minor')
# ax.xaxis.grid(True, linewidth = 0.85, color = 'red' ,which='minor')
ax.xaxis.grid(True)
ax.yaxis.grid(True)
# ax.grid(True, linestyle = '--', which = "minor")

a = ax.get_xgridlines()
b = a[5]
b.set_color('black')
b.set_linewidth(1)

fig.savefig('ETD_difference_function.eps', format='eps')



# a = ax.get_ygridlines()
# b = a[4]
# b.set_color('red')
# b.set_linewidth(1)
# plt.show()

```
