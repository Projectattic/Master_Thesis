```python
##### CHECKING CONTAINER LOCATIONS, DWELL TIMES #####

import stackClass
import containerClass
empty_stack = stackClass.Stack("A","1",[],6)

for x in range(1,empty_stack.stack_height+1):
    container_name = "ISO_"+str(x)
    empty_stack.add_container(containerClass.Container(container_name,float(x)))

#1
containers = empty_stack.get_containers()
print("1 " + str(containers))

#2
str_build = ""
for position in reversed(list(containers.keys())):
    str_build += position + " | " + str(containers[position]) + " | " + str(containers[position].get_location())+ " | " + str(containers[position].get_dwell_time())+"\n"
print(str_build)
```
