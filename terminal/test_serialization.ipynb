```python
from containerClass import Container
from stackClass import Stack
import containerClass
import json
import ast
from copy import deepcopy

empty_stack = Stack("A","1",[],6)
empty_stack.add_container(Container("ISO_123",float(1)))

def serialize(stack):
    stack_dict = deepcopy(stack.__dict__)
    # print(type(stack_dict['containers']))
    container_string = "{"
    for container_loc,container in stack_dict['containers'].items():
        # print(container)
        if container!= None:
            container_string+="\""+container_loc+"\": "+json.dumps(container.__dict__)+", "
        else:
            container_string+="\""+container_loc+"\": None, "
    container_string = container_string[:-2] #remove last comma
    container_string+="}"

    stack_dict['containers'] = json.loads(json.dumps(ast.literal_eval(container_string)))

    return json.dumps(stack_dict)

#check whether altering stack_dict changes values
print(empty_stack.__dict__)
print()
serialized_stack = serialize(empty_stack)

def deserialize(stack_string):
    deserialized = json.loads(stack_string)
    new_stack = Stack("","",[],0)
    for key,item in deserialized.items():
        new_stack.__dict__[key] = item

    temp_container_dict = {}

    for spot,container_dict in deserialized['containers'].items():

        if container_dict != None:
            temp_container_dict[spot] = Container(container_dict['container_number'],container_dict['dwell_time'])
        else:
            temp_container_dict[spot] = None
            
    new_stack.set_containers(temp_container_dict)

    return new_stack

deserialized_stack = deserialize(serialized_stack)

print(deserialized_stack.__dict__)
```

```python
########## deserializing and Serializing Row #######
import rowClass
import stackClass
import containerClass
from copy import deepcopy
import ast
import json 

def serialize(row):
    row_dict = deepcopy(row.__dict__)
    # print(type(stack_dict['containers']))
    stack_string = "{"
    for stack_loc,stack in row_dict['stacks'].items():
        # print(container)
        if stack!= None:
            stack_string+="\""+stack_loc+"\": "+stackClass.serialize(stack)+", "
        else:
            stack_string+="\""+stack_loc+"\": None, "
    stack_string = stack_string[:-2] #remove last comma
    stack_string+="}"

    # row_dict['stacks'] = stack_string
    row_dict['stacks'] = json.loads(stack_string)

    return json.dumps(row_dict)

test_row = rowClass.Row("A",[stackClass.Stack("A","1",[containerClass.Container("ISO_1",float(1)),containerClass.Container("ISO_2",float(2)),containerClass.Container("ISO_3",float(3)),containerClass.Container("ISO_4",float(4)),containerClass.Container("ISO_5",float(5)),containerClass.Container("ISO_6",float(6))],6),
stackClass.Stack("A","2",[containerClass.Container("ISO_7",float(7)),containerClass.Container("ISO_8",float(8)),containerClass.Container("ISO_9",float(9))],6),
stackClass.Stack("A","3",[containerClass.Container("ISO_10",float(10))],6)],5,6)
print(test_row)
serialized_row = serialize(test_row)
print(serialized_row)

def deserialize(row_string):
    deserialized = json.loads(row_string)
    new_row = rowClass.Row("",[],0,0)
    for key,item in deserialized.items():
        new_row.__dict__[key] = item

    temp_stack_dict = {}

    for spot,stack_dict in deserialized['stacks'].items():
        if stack_dict != None:
            temp_stack_dict[spot] = stackClass.deserialize_dict(stack_dict)
        else:
            temp_stack_dict[spot] = None
            
    new_row.set_stacks(temp_stack_dict)

    return new_row

deserialized_row = deserialize(serialized_row)
print(deserialized_row)
```
